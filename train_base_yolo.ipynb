```python
# ============================================
# 2. ИМПОРТ НЕОБХОДИМЫХ БИБЛИОТЕК
# ============================================

import numpy as np
import warnings

from ultralytics import YOLO

import torch

warnings.filterwarnings("ignore")
```

```python
# ============================================
# 3. ИНИЦИАЛИЗАЦИЯ МОДЕЛИ YOLOv8n
# ============================================

model = YOLO("models/YOLOv8n_base/yolov8n.pt")
```

```python
results = model.train(
    data='models/YOLOv8n_base/dataset_original.yaml',
    epochs=10,
    batch=8,
    imgsz=640,
    device='0',  # 'cpu' или '0' для GPU
    name='yolov8_base',
    optimizer='Adam',
    lr0=0.000001,
    augment=True,  # автоматические аугментации
    pretrained=True
)
```
